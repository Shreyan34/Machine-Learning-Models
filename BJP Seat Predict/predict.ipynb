```python
#importing modules
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
#main model

predict_year = int(input('Enter the year of prediction: '))
#taking input of the model

data = pd.read_csv('Seats VS Years (CSV).csv') #reading CSV file

#plotting the data
plt.plot(data['Years'], data['Seats'])
plt.title('The Model So Far')
plt.xlabel('Years')
plt.ylabel('Seats Won')
plt.show()

#creating the model
model = LinearRegression()
model.fit(data[['Years']], data[['Seats']]) #feeding data to the model

print('In the year:', predict_year, ', BJP shall get: ', model.predict([[predict_year]]), 'seats') #predicting the outcome
```
